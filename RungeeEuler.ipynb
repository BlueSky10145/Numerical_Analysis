```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return x + y

x0, y0 = 0, 1
h = 0.5
n = 4

x_euler = [x0]
y_euler = [y0]
x, y = x0, y0
for _ in range(n):
    y_new = y + h * f(x, y)
    x_new = x + h
    x_euler.append(x_new)
    y_euler.append(y_new)
    x, y = x_new, y_new

x_rk = [x0]
y_rk = [y0]
x, y = x0, y0
for _ in range(n):
    k1 = h * f(x, y)
    k2 = h * f(x + h/2, y + k1/2)
    k3 = h * f(x + h/2, y + k2/2)
    k4 = h * f(x + h, y + k3)

    y_new = y + (k1 + 2*k2 + 2*k3 + k4) / 6
    x_new = x + h
    x_rk.append(x_new)
    y_rk.append(y_new)
    x, y = x_new, y_new

x_exact = np.linspace(0, 2, 100)
y_exact = 2 * np.exp(x_exact) - x_exact - 1

plt.figure(figsize=(9, 6))
plt.plot(x_euler, y_euler, label="Euler Method", marker="o", linestyle="-", color="red")
plt.plot(x_rk, y_rk, label="Runge-Kutta (RK4)", marker="s", linestyle="-", color="blue")
plt.plot(x_exact, y_exact, label="Exact Solution", linestyle="dashed", color="black")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graphical Derivation of Runge-Kutta Equations")
plt.legend()
plt.grid()
plt.show()
```

## Euler method

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return x + y

x0, y0 = 0, 1
h = 0.5
n = 4

x_vals = [x0]
y_vals = [y0]

x, y = x0, y0
for _ in range(n):
    y = y + h * f(x, y)
    x = x + h
    x_vals.append(x)
    y_vals.append(y)

plt.plot(x_vals, y_vals, 'o-', label='Euler Method', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Euler Method')
plt.grid()
plt.legend()
plt.show()
```

## Rungee Kutta 

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return x + y

x0, y0 = 0, 1
h = 0.5
n = 4

x_vals = [x0]
y_vals = [y0]

x, y = x0, y0
for _ in range(n):
    k1 = h * f(x, y)
    k2 = h * f(x + h/2, y + k1/2)
    k3 = h * f(x + h/2, y + k2/2)
    k4 = h * f(x + h, y + k3)
    y = y + (k1 + 2*k2 + 2*k3 + k4) / 6
    x = x + h
    x_vals.append(x)
    y_vals.append(y)

plt.plot(x_vals, y_vals, 's-', label='Runge-Kutta 4th Order', color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Runge-Kutta 4th Order Method')
plt.grid()
plt.legend()
plt.show()
```

## simpson

```python
import numpy as np

def f(x):
    return np.sin(x)  # Example function

def simpson_rule(a, b, n):
    if n % 2 != 0:
        raise ValueError("Number of intervals n must be even.")
    
    h = (b - a) / n
    x = np.linspace(a, b, n + 1)
    y = f(x)
    
    S = y[0] + y[-1] + 4 * sum(y[1:-1:2]) + 2 * sum(y[2:-2:2])
    return (h / 3) * S

a, b = 0, np.pi
n = 10  # Must be even
result = simpson_rule(a, b, n)
print(f"Integral of sin(x) from {a} to {b}  {result}")
```

## Tailor

```python
import numpy as np
import math

def taylor_series_exp(x, n_terms):
    result = 0
    for n in range(n_terms):
        result += (x**n) / math.factorial(n)
    return result

x_val = 1.0
n_terms = 10

approx = taylor_series_exp(x_val, n_terms)
exact = np.exp(x_val)

abs_error = abs(exact - approx)
rel_error = abs_error / abs(exact) * 100  # percentage relative error

print(f"Taylor series approximation of e^{x_val} using {n_terms} terms: {approx}")
print(f"Exact value: {exact}")
print(f"Absolute error: {abs_error}")
print(f"Relative error (%): {rel_error}")
```

## trapezoidal

```python
import numpy as np

def trapezoidal_rule(f, a, b, n):
    h = (b - a) / n
    s = 0.5 * (f(a) + f(b))
    for i in range(1, n):
        s += f(a + i * h)
    return h * s

# Example function
def f(x):
    return x**2

a = 0
b = 2
n = 10

result = trapezoidal_rule(f, a, b, n)
print(f"Approximate integral using Trapezoidal Rule: {result}")
```
